## Basic Statistics

Let's compute some basic statistics about the dataset.

```python
# Check class distribution in training set
train_class_counts = train_df['label'].value_counts()
print("Class distribution in training set:")
print(train_class_counts)
print(f"Class balance: {train_class_counts[0] / train_class_counts[1]:.2f}")
```

```python
# Plot class distribution
fig, ax = plt.subplots(figsize=(10, 6))
sns.countplot(x='label', data=train_df, ax=ax)
ax.set_xticklabels(['Negative', 'Positive'])
ax.set_title('Class Distribution in Training Set')
ax.set_xlabel('Sentiment')
ax.set_ylabel('Count')

# Add count labels
for i, count in enumerate(train_class_counts):
    ax.text(i, count + 100, f"{count:,}", ha='center')

plt.tight_layout()
save_figure(fig, 'class_distribution.png', visualizations_dir)
plt.show()
```

## Vocabulary Analysis

Let's analyze the vocabulary used in the reviews.

```python
# Download NLTK resources if needed
nltk.download('stopwords')
nltk.download('punkt')

# Get English stopwords
stop_words = set(stopwords.words('english'))
```

```python
# Function to preprocess text
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove special characters and digits
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenize
    tokens = text.split()
    # Remove stopwords
    tokens = [token for token in tokens if token not in stop_words]
    return tokens
```

```python
# Process a sample of reviews for efficiency
sample_size = 1000
train_sample = train_df.sample(sample_size, random_state=42)

# Preprocess text and count word frequencies
positive_tokens = []
negative_tokens = []

for _, row in train_sample.iterrows():
    tokens = preprocess_text(row['text'])
    if row['label'] == 1:  # Positive
        positive_tokens.extend(tokens)
    else:  # Negative
        negative_tokens.extend(tokens)

# Count word frequencies
positive_word_counts = Counter(positive_tokens)
negative_word_counts = Counter(negative_tokens)

print(f"Number of unique words in positive reviews: {len(positive_word_counts)}")
print(f"Number of unique words in negative reviews: {len(negative_word_counts)}")
```

```python
# Most common words in positive reviews
print("Most common words in positive reviews:")
for word, count in positive_word_counts.most_common(20):
    print(f"{word}: {count}")
```

```python
# Most common words in negative reviews
print("Most common words in negative reviews:")
for word, count in negative_word_counts.most_common(20):
    print(f"{word}: {count}")
```

```python
# Generate word cloud for positive reviews
positive_wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    max_words=100,
    contour_width=1,
    contour_color='steelblue'
).generate(' '.join(positive_tokens))

fig, ax = plt.subplots(figsize=(16, 8))
ax.imshow(positive_wordcloud, interpolation='bilinear')
ax.axis('off')
ax.set_title('Word Cloud - Positive Reviews', fontsize=20)
plt.tight_layout()
save_figure(fig, 'positive_wordcloud.png', visualizations_dir)
plt.show()
```

```python
# Generate word cloud for negative reviews
negative_wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    max_words=100,
    contour_width=1,
    contour_color='steelblue'
).generate(' '.join(negative_tokens))

fig, ax = plt.subplots(figsize=(16, 8))
ax.imshow(negative_wordcloud, interpolation='bilinear')
ax.axis('off')
ax.set_title('Word Cloud - Negative Reviews', fontsize=20)
plt.tight_layout()
save_figure(fig, 'negative_wordcloud.png', visualizations_dir)
plt.show()
```

## Distinctive Words Analysis

Let's find words that are more distinctive for each sentiment class.

```python
# Calculate word frequency ratios
word_ratio = {}
min_count = 5  # Minimum count to consider

for word in set(list(positive_word_counts.keys()) + list(negative_word_counts.keys())):
    pos_count = positive_word_counts.get(word, 0)
    neg_count = negative_word_counts.get(word, 0)
    
    # Skip words that are too rare
    if pos_count + neg_count < min_count:
        continue
    
    # Add a small constant to avoid division by zero
    ratio = (pos_count + 0.1) / (neg_count + 0.1)
    word_ratio[word] = ratio
```

```python
# Most distinctive positive words
most_positive = sorted(word_ratio.items(), key=lambda x: x[1], reverse=True)[:20]
print("Most distinctive positive words:")
for word, ratio in most_positive:
    pos_count = positive_word_counts.get(word, 0)
    neg_count = negative_word_counts.get(word, 0)
    print(f"{word}: {ratio:.2f} (Pos: {pos_count}, Neg: {neg_count})")
```

```python
# Most distinctive negative words
most_negative = sorted(word_ratio.items(), key=lambda x: x[1])[:20]
print("Most distinctive negative words:")
for word, ratio in most_negative:
    pos_count = positive_word_counts.get(word, 0)
    neg_count = negative_word_counts.get(word, 0)
    print(f"{word}: {1/ratio:.2f} (Pos: {pos_count}, Neg: {neg_count})")
```

```python
# Plot most distinctive words
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Positive words
pos_words = [word for word, _ in most_positive]
pos_ratios = [ratio for _, ratio in most_positive]
ax1.barh(pos_words[::-1], pos_ratios[::-1], color='green', alpha=0.7)
ax1.set_title('Most Distinctive Positive Words')
ax1.set_xlabel('Positive/Negative Ratio')

# Negative words
neg_words = [word for word, _ in most_negative]
neg_ratios = [1/ratio for _, ratio in most_negative]
ax2.barh(neg_words[::-1], neg_ratios[::-1], color='red', alpha=0.7)
ax2.set_title('Most Distinctive Negative Words')
ax2.set_xlabel('Negative/Positive Ratio')

plt.tight_layout()
save_figure(fig, 'distinctive_words.png', visualizations_dir)
plt.show()
```

## BERT Tokenization Analysis

Let's analyze how BERT tokenizes the reviews to understand the input to our model.

```python
from transformers import AutoTokenizer

# Load BERT tokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Tokenize a sample review
sample_review = train_df['text'].iloc[0]
print(f"Sample review:\n{sample_review[:500]}...\n")

# Tokenize
tokens = tokenizer.tokenize(sample_review)
print(f"Number of tokens: {len(tokens)}")
print(f"First 50 tokens: {tokens[:50]}")
```

```python
# Analyze token lengths for a sample of reviews
token_lengths = []
for text in train_df['text'].sample(1000, random_state=42):
    tokens = tokenizer.tokenize(text)
    token_lengths.append(len(tokens))

# Plot token length distribution
fig, ax = plt.subplots(figsize=(12, 6))
sns.histplot(token_lengths, bins=50, kde=True, ax=ax)
ax.set_title('Distribution of BERT Token Lengths')
ax.set_xlabel('Number of Tokens')
ax.set_ylabel('Frequency')

# Add vertical line at BERT's max sequence length
ax.axvline(x=512, color='red', linestyle='--', label='BERT Max Length (512)')
ax.legend()

# Add statistics
stats_text = (
    f"Mean: {np.mean(token_lengths):.1f} tokens\n"
    f"Median: {np.median(token_lengths):.1f} tokens\n"
    f"Max: {np.max(token_lengths)} tokens\n"
    f"Reviews > 512 tokens: {sum(l > 512 for l in token_lengths)} ({sum(l > 512 for l in token_lengths)/len(token_lengths)*100:.1f}%)"
)
ax.text(
    0.95, 0.95, stats_text,
    transform=ax.transAxes,
    verticalalignment='top',
    horizontalalignment='right',
    bbox=dict(boxstyle='round', facecolor='white', alpha=0.8)
)

plt.tight_layout()
save_figure(fig, 'token_length_distribution.png', visualizations_dir)
plt.show()
```

## Conclusion

In this notebook, we've performed exploratory data analysis on the IMDB Movie Review dataset. Here are the key findings:

1. The dataset is balanced with an equal number of positive and negative reviews.
2. The reviews vary in length, with some being quite long.
3. There are distinctive words that are more common in positive and negative reviews.
4. When tokenized with BERT, some reviews exceed the maximum sequence length of 512 tokens, which means we'll need to truncate them.

These insights will help us in building and fine-tuning our BERT-based sentiment analysis model.
